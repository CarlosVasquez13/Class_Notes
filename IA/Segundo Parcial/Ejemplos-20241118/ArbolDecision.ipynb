```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import  train_test_split
from sklearn.metrics import accuracy_score

from sklearn.datasets import load_iris
from sklearn import tree
```

```python
data = load_iris()
```

```python
print(data)
```

```python
X = data.data
y = data.target
```

```python
print(X)
```

```python
print(y)
```

```python
print('Objetivos distintos: ', np.unique(y))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.5, random_state=42)
```

```python
arbol_clasificador = DecisionTreeClassifier(criterion='gini', max_depth=5, random_state = 43 ) #entropy
arbol_clasificador.fit(X_train, y_train)
```

```python
y_pred = arbol_clasificador.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print('Precisión del modelo: ', accuracy)
```

```python
tree.plot_tree(arbol_clasificador)
```
