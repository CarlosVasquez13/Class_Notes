```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_cobros = pd.read_csv('insurance.csv')
df_cobros.head()
```

```python

# Crear un boxplot para visualizar los cuantiles de la columna "charges"
df_cobros.boxplot(column=['charges'], figsize=(10, 6))
plt.title('Boxplot de los cuantiles de charges en df_cobros')
plt.ylabel('Valores')
plt.show()
```

```python
first_quantile = df_cobros['charges'].quantile(0.25)
third_quantile = df_cobros['charges'].quantile(0.75)
iqr = third_quantile - first_quantile
```

```python
inferior_limit = first_quantile - (1.5 * iqr)
superior_limit = third_quantile + (1.5 * iqr)
```

```python
df_without_outliers = df_cobros[(df_cobros['charges'] >= inferior_limit) & (df_cobros['charges'] <= superior_limit)]
df_outliers = df_cobros[(df_cobros['charges'] < inferior_limit) | (df_cobros['charges'] > superior_limit)]
df_outliers.boxplot(column=['charges'], figsize=(10, 6))
plt.title('Boxplot de los cuantiles de charges en df_without_outliers')
plt.ylabel('Valores')
plt.show()
```
